# Pre-Modeling Transformation Notebook


## Section 1 – Project Alignment

### Research Questions

1. What property characteristics (price, size, age, and features) influence whether a listing becomes stale (DOM > 60)?

2. How does pricing relative to the local market (Expectation Gap) affect the likelihood of a listing becoming stale?

3. How do location and neighborhood factors (distance to downtown, community characteristics, and income levels) impact listing performance?

### Type of Analysis

This project uses classification modeling to predict whether a listing becomes stale.

The following models will be evaluated:

- Logistic Regression
- Random Forest
- Gradient Boosting

To ensure robust model performance, the following techniques will be applied:

- Train/test split to evaluate generalization
- Cross-validation to reduce overfitting
- Hyperparameter tuning to optimize model performance
- Class imbalance handling to improve detection of stale listings

## Section 2 – Advanced Feature Engineering

### Feature 1: Price per Square Foot

- Variable: price_per_sqft
- Measures property value relative to size
- Helps identify overpriced listings

### Feature 2: Price per Bedroom

- Variable: price_per_bedroom
- Measures cost efficiency of space
